```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML/online.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
x = df.drop(['Profit'],axis = 1)
x
```

```python
y = df['Profit']
y
```

```python
city = pd.get_dummies(df['Area'])
city
```

```python
merged = pd.concat([df,city],axis=1)
merged
```

```python
 final = merged.drop(['Area'],axis=1)
 final
```

```python
x = final.drop(['Profit'],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(xtrain,ytrain)
```

```python
ytest
```

```python
pred = reg.predict(xtest)
pred
```

```python
reg.score(xtest,ytest)
```

```python
from sklearn.metrics import r2_score
score = r2_score(ytest,pred)
score
```

```python
diff = pd.DataFrame({'Actual Value':ytest,'Predict Value':pred,'Diffirence':ytest-pred})
diff
```
